HISTOGRAM


```python
# To plot a histogram we need all the numerical columns
numerical_cols = df.select_dtypes(include=['int','float']).columns.to_list() # convert to python list
# print(numerical_cols)
# remove the id column
numerical_cols.remove('id')
df[numerical_cols].hist(bins=20, figsize=(10,15))
plt.tight_layout()
```

Heatmap

```python
plt.figure(figsize=(12, 6))
sns.heatmap(df[numerical_cols].corr(), annot=True, cmap='coolwarm', center=0, fmt='.2f')
plt.title('Correlation Heatmap of Numerical Variables')
plt.show()
```
